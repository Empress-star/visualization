```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

#creating data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
#setting dark mode
plt.style.use('dark_background')

#creating line plot with two series

fig, ax = plt.subplots(figsize = ( 5,3))
ax.plot(x, y1)
ax.plot (x,y2)
```

```python
#labeling series and adding a legend

fig, ax = plt.subplots(figsize = ( 5,3))
ax.plot(x, y1, label = "Person 1")
ax.plot (x,y2, label = "Person 2")
ax.legend(loc = 'lower right')
```

```python
#adding frame and shadow to legend

fig, ax = plt.subplots(figsize = ( 5,3))
ax.plot(x, y1, label = "Person 1")
ax.plot (x,y2, label = "Person 2")
ax.legend(loc = 'lower right',
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True)
```

```python
# moving legend outside of plot area

fig, ax = plt.subplots(figsize = ( 5,3))
ax.plot(x, y1, label = "Person 1")
ax.plot (x,y2, label = "Person 2")
ax.legend(loc = 'upper left', # what part of the box we are anchoring
          bbox_to_anchor = (1,1)) # anchor is where we are anchoring our legend
```

```python
# adding anotations with the text() function

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = 'lower right')
ax.text(10,95, 'This value is important!')
```

```python
# modifying text annotation

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x, y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc = 'lower right')
ax.text(10,95, 'This value is important!',
        ha = 'center',
        color = 'red',
        size = 20)
```

```python
# positioning text annotations

#creating 10 by 10 empty axis
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

#creating text with transformations
ax.text(1,5,". Data: (1,5)",transform = ax.transData) #setting it to scale to Data
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform = ax.transAxes) #setting it to scale to proportion of axes object
ax.text(0.2,0.2, ". Figure: (0.2, 0.2)", transform = fig.transFigure)# settng it to scale to proportion to Figure object
```

```python
#subplots

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y2 = np.array([110,180,240,99,220])
```

```python
# setting to two plots next to each other

fig, (ax1, ax2) = plt.subplots(ncols = 2,  #ncols is columns
                               nrows = 1,
                               figsize = (7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)

fig.show()
```

```python
# analyzing codes

font1 = {'family': 'serif', 'size': 10}
font2 = {'family': 'serif', 'size': 14}

fig, (ax1, ax2) = plt.subplots(ncols = 2,  #ncols is columns
                               nrows = 1,
                               figsize = (7,3))
ax1.scatter(x1,y1,
            marker = 'o',
            color = 'red')
ax2.bar(x2,y2, color = 'darkgreen')

ax1.set_title('Scatterplot', fontdict = font2)
ax1.set_ylabel('Measure', fontdict = font1)
ax1.set_xlabel('Values', fontdict = font1)
ax2.set_title('One Piece Characters', fontdict = font2)
ax2.set_ylabel('Strength', fontdict = font1)
ax2.set_xlabel('Pirate', fontdict = font1)

fig.show()
```

```python
# someaxes and mosaics

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize = (7,4))

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
```

```python
#adding labels - this is bad

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize = (7,4))

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```

```python
# layouts

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize = (7,4),
                                   layout = 'constrained') #layout to make things fit

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)

someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 18)
```
