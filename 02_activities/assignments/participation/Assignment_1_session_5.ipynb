```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# create data

x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

# define figures and axes

fig, ax = plt.subplots(figsize = (7,3))

# set dark mode

plt.style.use('dark_background')

# plot bot and plot to ax

ax.bar(x, y1,
       color = 'indigo')
ax.plot(x, y2,
        color = 'red')
```

```python
# calculate standard deviation of data

y2_sd = np.std(y2) #standard deviation from numpy

# plot line as before

fig, ax = plt.subplots(figsize = (7,3))

ax.plot(x,y2,
        color = 'red')
```

```python
# add error information using errorbar() - adding some form of error mesaurement is standard practice
fig, ax = plt.subplots(figsize = (7,3))

ax.plot(x,y2,
        color = 'red')


ax.errorbar(x, # our x values
            y2, #our y value
            yerr = y2_sd, # specifies that were plotting vertical error bars
            fmt = 'none') # makes sure we're not plotting the actual data points, only the error
```

```python
# customizing error bar appearance

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2,
        color = 'red')
ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor = 'indigo', #error color
            elinewidth = 4, # error linewidth
            capsize = 6, # creates caps to the bars
            capthick = 4)
```

```python
# errorevery for customizing intervals for the error lines
fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x,y2,
        color = 'red')
ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt = 'none',
            ecolor = 'indigo', #error color
            elinewidth = 4, # error linewidth
            capsize = 6, # creates caps to the bars
            capthick = 4,
            errorevery = 2)
```

# Adding images to plots

```python
# using libraries for images from the internet
from PIL import Image #to open images
import requests #to get images from URLs
from io import BytesIO # to store images
```

```python
# get image of Luffy

response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# make basic line plot

fig, ax = plt.subplots(figsize = (7,3))
ax.plot(x, y2,
        color = 'red')

# create new axis as overlay for figure as a container for the image
ax_image = fig.add_axes([0.1, # x coordinate on figure
                         0.11, # y coordinate
                         0.15, # image width
                         0.35] # image height
                        )

# use imshow() to add image
ax_image.imshow(image)
ax_image.axis('off')
```
